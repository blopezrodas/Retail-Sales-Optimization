# Retail Sales Optimization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# load data
data = pd.read_csv('SuperStore.csv', encoding='ISO-8859-1')
data.head()
```

```python
# data description
data.info()
```

```python
# drop row id column
data.drop(columns='Row ID', inplace = True)
```

```python
# convert date columns to datetime type
data['Order Date'] = pd.to_datetime(data['Order Date'], errors = 'coerce')
data['Ship Date'] = pd.to_datetime(data['Ship Date'], errors = 'coerce')
print(data['Order Date'].dtype)
print(data['Ship Date'].dtype)
```

```python
# check if there were any invalid dates after conversion
data.isnull().sum()
```

```python
# plot numeric columns
data_numeric = data.select_dtypes(include='number')
sns.pairplot(data_numeric)
```

```python
data.hist(figsize=(12, 10), bins=30)
plt.tight_layout()
plt.show()
```

```python
data.to_csv('SuperStore_cleaned.csv', index = False)
```
